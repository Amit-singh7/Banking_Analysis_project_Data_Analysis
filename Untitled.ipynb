```python
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Mid', 'High']

df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels,right=False)
```

```python
df['Income Band'].value_counts()
```

```python
categorical_cols = df[['BRId', 'GenderId', 'IAId', 'Amount of Credit Cards', 'Nationality', 'Occupation', 'Fee Structure', 'Loyalty Classification', 'Properties Owned', 'Risk Weighting', 'Income Band']].columns

for col in categorical_cols:
    print(f"value counts for' {col}':")
    display(df[col].value_counts())
```

```python
#univarient analysis
for i, predicator in enumerate(df[['BRId', 'GenderId', 'IAId', 'Amount of Credit Cards', 'Nationality', 'Occupation', 'Fee Structure', 'Loyalty Classification', 'Properties Owned', 'Risk Weighting', 'Income Band']].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predicator, hue="GenderId")
```

```python
#Bivarient analysis
for i, predicator in enumerate(df[['BRId', 'GenderId', 'IAId', 'Amount of Credit Cards', 'Nationality', 'Occupation', 'Fee Structure', 'Loyalty Classification', 'Properties Owned', 'Risk Weighting', 'Income Band']].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predicator)
```

```python
#Hist plot

for col in categorical_cols:
    if col =="Occupation":
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title("Histogram of Occupation count")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
numerical_col=['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

#univariate analysis % visualization
plt.figure(figsize=(15,10))
for i, col in enumerate(numerical_col):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
plt.show()     
```

```python
#heatmaps

correlation_matrix = df[numerical_col].corr()

plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```
